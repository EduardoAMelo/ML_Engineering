```python
import pandas as pd
```

```python
data = pd.read_csv('data/imigrantes_canada.csv', sep=',')
data.head()
```

```python
data.set_index('País', inplace=True)
data
```

```python
anos = list(map(str, range(1980, 2014)))
brasil = data.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

## Matplotlib

```python
import matplotlib.pyplot   as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

### Melhorando visualização do gráfico

```python
# x width, y height in pol
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# Show just the selected x descriptions
plt.title('Imigração do Brasil para o Canada')
plt.xticks([str(x) for x in range(1980, 2014, 5)])

# x and y labels
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
# show graph
plt.show()
```

### Subplots

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigrantes do Brasil para o Canada', fontsize=18)
# Much more easy than xticks
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize=(12))
ax.yaxis.set_tick_params(labelsize=(12))
ax.set_xlabel('Ano', fontsize=(14))
ax.set_ylabel('Imigrantes', fontsize=(14))

plt.show()
```

### Dois gráficos ao mesmo tempo

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigrantes do Brasil para o Canada')
# Much more easy than xticks
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title(' Boxplot Imigrantes do Brasil para o Canada')
# Much more easy than xticks
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,7))
fig.suptitle('Imigração dos 4 Países')
fig.subplots_adjust(hspace=0.5, wspace=0.4)

axs[0, 0].plot(data.loc['Brasil', anos],lw=3, marker='o')
axs[0, 0].set_title('Brasil', loc='left')

axs[0, 1].plot(data.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(data.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(data.loc['Peru', anos])
axs[1, 1].set_title('Peru')

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')
plt.show()
```

```python
america_sul = data.query('Região=="América do Sul"')
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
america_sul_sorted
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')


fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)

for i, v, in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = data.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10, x=top_10.index, y='Total')
```

```python
sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
ax.set(title='Países com maior imigração para o Canada',
       xlabel='Número de imigrantes',
       ylabel='')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
ax.set_title('Países com maior imigração para o Canada',loc='left')
ax.set_xlabel('Número de imigrantes')
```

```python
def gerar(pallete):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
    ax.set_title('Países com maior imigração para o Canada',loc='left')
    ax.set_xlabel('Número de imigrantes')
    plt.show()
```

```python
gerar('Blues')
```
