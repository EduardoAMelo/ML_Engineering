```python
import pandas as pd
```

```python
data = pd.read_csv('data/imigrantes_canada.csv', sep=',')
data.head()
```

```python
data.set_index('País', inplace=True)
data
```

```python
anos = list(map(str, range(1980, 2014)))
brasil = data.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot   as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

### Melhorando visualização do gráfico

```python
# x width, y height in pol
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# Show just the selected x descriptions
plt.title('Imigração do Brasil para o Canada')
plt.xticks([str(x) for x in range(1980, 2014, 5)])

# x and y labels
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
# show graph
plt.show()
```

### Subplots

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigrantes do Brasil para o Canada', fontsize=18)
# Much more easy than xticks
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize=(12))
ax.yaxis.set_tick_params(labelsize=(12))
ax.set_xlabel('Ano', fontsize=(14))
ax.set_ylabel('Imigrantes', fontsize=(14))

plt.show()
```

### Dois gráficos ao mesmo tempo

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigrantes do Brasil para o Canada')
# Much more easy than xticks
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title(' Boxplot Imigrantes do Brasil para o Canada')
# Much more easy than xticks
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,7))
fig.suptitle('Imigração dos 4 Países')
fig.subplots_adjust(hspace=0.5, wspace=0.4)

axs[0, 0].plot(data.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(data.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(data.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(data.loc['Peru', anos])
axs[1, 1].set_title('Peru')

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')
plt.show()
```
